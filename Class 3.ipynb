```python
# from scratch implementation of Singly Linked Lists

class SinglyLinkedList:
    class _Node:
        def __init__(self, datum):
            self.data = datum
            self.next = None
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, datum):
        new_node = self._Node(datum)
        self.size += 1
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def remove(self, datum):
        # removes the first instance of 'datum' from the list if found; ValueError if not found
        if self.head:
            previous = None
            current = self.head
            while current and current.data != datum:
                previous = current
                current = current.next
            if not current:
                raise ValueError("%s not found in list" % datum)
            else:
                if previous:
                    previous.next = current.next
                else:
                    self.head = self.head.next
                    if not self.head:
                        self.tail = None
                self.size -= 1
        else:
            raise ValueError("Target value %s is not in list" % datum)

    def insert(self, index, datum):
        # This method should insert a new node with 'datum' before the target index
         if index < 0 or index > self.size:
            raise IndexError("Index out of bounds")
        
         new_node = self._Node(datum)
         if index == 0:
             new_node.next = self.head
             self.head = new_node
             if self.size == 0:
                 self.tail = new_node
         else:
             current = self.head
             for _ in range(index - 1):
                 current = current.next
             new_node.next = current.next
             current.next = new_node
             if new_node.next is None:
                 self.tail = new_node
         self.size += 1

    def count(self, datum):
        count = 0
        current = self.head
        while current:
            if current.data == datum:
                count += 1
            current = current.next
        return count

    def reverse(self):
        previous = None
        current = self.head
        self.tail = self.head  
        while current:
            next_node = current.next 
            current.next = previous  
            previous = current       
            current = next_node      
        self.head = previous
            
    def __len__(self):
        return self.size

    def __str__(self):
        out = "["
        if self.head:
            current = self.head
            out += "%s" % repr(current.data)
            current = current.next
            while current:
                out += ", %s" % repr(current.data)
                current = current.next
        out += "]"
        return out
                    
```

```python
sll = SinglyLinkedList()

for number in range(1, 11):
    sll.append(number)

print(" Our list: %s" % sll)
```
